# **MTCNN Face Detection**

```python
import cv2
from mtcnn import MTCNN

# Use raw string (prefix with r) to avoid escaping issues
image_path = r'C:\Users\rosha\OneDrive\Desktop\xebia intern\face-clustering-main\sample_images\image 1.jpg'

# Load an image using OpenCV
image = cv2.imread(image_path)

if image is None:
    raise ValueError(f"Could not load image from path: {image_path}")

# Convert the image from BGR to RGB format (if necessary)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Only if using matplotlib for display

# Initialize MTCNN detector
detector = MTCNN()

# Detect faces in the image
faces = detector.detect_faces(image_rgb)

# Draw bounding boxes around detected faces (if needed)
for face in faces:
    x, y, width, height = face['box']
    cv2.rectangle(image_rgb, (x, y), (x + width, y + height), (255, 0, 0), 2)

# Display the image with detected faces (if using matplotlib)
import matplotlib.pyplot as plt
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
import cv2
from mtcnn import MTCNN

# Define the image path
image_path = r'C:\Users\rosha\OneDrive\Desktop\xebia intern\face-clustering-main\sample_images\image 1.jpg'

# Load an image using OpenCV
img = cv2.imread(image_path)

# Check if the image was loaded correctly
if img is None:
    raise ValueError(f"Image not loaded correctly from path: {image_path}")

# Convert the image from BGR to RGB format (required for MTCNN and matplotlib)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Initialize MTCNN detector
detector = MTCNN()

# Detect faces in the image
result = detector.detect_faces(img_rgb)

# Print the detected faces and their confidence
print(result)

if result:
    bounding_box = result[0]['box']
    keypoints = result[0]['keypoints']

    # Draw bounding box
    cv2.rectangle(img,
                  (bounding_box[0], bounding_box[1]),
                  (bounding_box[0]+bounding_box[2], bounding_box[1]+bounding_box[3]),
                  (255, 0, 0), 2)

    # Draw keypoints
    for point_name in keypoints:
        cv2.circle(img, keypoints[point_name], 2, (0, 0, 255), 2)

    # Display the image with bounding boxes and keypoints
    cv2.imshow('Detected Faces', cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("No faces detected.")


```
